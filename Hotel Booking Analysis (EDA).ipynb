```python
#IMPORTING LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#LOADING THE DATASET
df= pd.read_csv('hotel_booking.csv')
```

```python
#EXPLORATORY DATA ANALYSIS AND DATA CLEANING

#Too check top 5 Rows of the dataset
df.head()
```

```python
#To check bottom 5 rows of the dataset
df.tail()
```

```python
#To check shape of the dataset
df.shape
```

```python
#To check columns
df.columns
```

```python
#To check columns datatype
df.info()
```

```python
#Some columns are irrelevant having personal info like(name,email,phone number,credit-card) so we will drop these columns
df.drop(['name','email','phone-number','credit_card'] ,axis=1, inplace=True)
```

```python
# Now the dataset has only general information and no personal details
df.columns
```

```python
#Converting reservation_status_date columns dataype from object to date
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
#To check categorical columns(object datatype) unique/null values
df.describe(include='object')
```

```python
#To check values coming under these categories
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
#Checking missing values
df.isnull().sum()
```

```python
#Since agents & company columns has huge missing/null values, so we will delete both the columns itself 
#And we will also drop null values of children & country columns
df.drop(['agent','company'], axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#To check summary statistics of the numerical columns
df.describe()
```

```python
#REMOVING THE OUTLIERS
#The dataset has many outliers so we will remove all of these

#Checking outliers in children column

df['children'].plot(kind='box')
```

```python
#Removing the outliers in children column
df=df[df['children']<6]
```

```python
#Checking after removing
df['children'].plot(kind='box')
```

```python
#Checking outliers in babies column

df['babies'].plot(kind='box')
```

```python
#Removing the ouliers
df=df[df['babies']<8]
```

```python
#Checking after removing

df['babies'].plot(kind='box')
```

```python
#Removing the outliers of 'adr'(i.e price) 
df=df[df['adr']<5000]

#Removing values below 0
df=df[df['adr']>0]
```

```python
#checking after removing
df['adr'].plot(kind='box')
```

```python
#Now the dataset is cleaned

df.describe()
```

```python
#NOW WE WILL PERFORM ANALYSIS AND VISUALIZATIONS

#Checking the amount of cancelled reservation and non-cancelled reservation from (2015 to 2017)
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k', width=0.7)
plt.show()
```

```python
#Check the cancelation rate and non-cancellation rate of both City Hotels & Resort Hotels
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in different Hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
ax1.legend(['Not Canceled','Canceled'])
plt.show()
```

```python
#To check the percentage of canceled reservations and non-canceled reservations of Resort Hotel
resort_hotel= df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#To check the percentage of canceled reservations and non-canceled reservations of City Hotel
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Does price affect the cancellation of Resort Hotels 
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()

#Does price affect the cancellation of City Hotels 
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#Lets Visualize this
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in Resort and City Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#To check which month has more reservations as well as cancellations
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
ax1.legend(bbox_to_anchor=(1,1))
plt.legend(['Not Canceled','Canceled'],fontsize=20)
plt.title('Reservation Status per Month',fontsize=25)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Number of Reservations', fontsize=15)
plt.show()
```

```python
#To check Average Daily Rate of each month
plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize=30)
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')['adr'].sum().reset_index())
plt.xlabel('Month', fontsize=15)
plt.ylabel('ADR', fontsize=15)
plt.show()      
```

```python
#To check the cancellation rate on the basis of country
#Top 10 countries
canceled_data = df[df['is_canceled']==1]
top_10_countries=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with Canceled Reservation',fontsize=20)
plt.pie(top_10_countries,autopct='%.2f',labels= top_10_countries.index)
plt.show()
```

```python
#To check from where the clients are coming
df['market_segment'].value_counts()
```

```python
#To check percentage counts
df['market_segment'].value_counts(normalize=True)
```

```python
#To check the percentage of cancellations are coming from where
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
#To check the Average Daily Rate(price) of Canceled reservations are high or non-Canceled reservations
canceled_data_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_data_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label='canceled')
plt.legend(fontsize=15)
plt.show()
```

```python
#We will filter the data from 2016 to sept 2017 for both canceled & not_cancele reservations (since the above fig is messy)
canceled_data_adr=canceled_data_adr[(canceled_data_adr['reservation_status_date']>'2016')&(canceled_data_adr['reservation_status_date']<'2017-09')]
not_canceled_data_adr=not_canceled_data_adr[(not_canceled_data_adr['reservation_status_date']>'2016')&(canceled_data_adr['reservation_status_date']<'2017-09')]

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label='canceled')
plt.legend(fontsize=15)
plt.show()
```
